## 1. Introduction to PCA

**Principal Component Analysis (PCA)** is an unsupervised dimensionality reduction technique that transforms high-dimensional data into lower-dimensional space while preserving maximum variance.[web:6][web:8]

### Key Concepts:
- **Principal Components (PCs)**: New orthogonal axes that capture maximum variance
- **Variance Explained**: First PC captures most variance, subsequent PCs capture decreasing variance
- **Dimensionality Reduction**: Reduce features from p to k (k << p) dimensions



```python
### Demo: Iris Dataset

# importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns # for visualization purposes
from sklearn.datasets import load_iris # Iris Dataset
from sklearn.preprocessing import StandardScaler # sklearn Python library
from sklearn.decomposition import PCA
```

```python
### Using Iris Dataset ### 

# Load data

iris = load_iris()

X = iris.data # samples or observations
Y = iris.target # features values (columns)

print(f"Original shape: {X.shape}")
print(f"Features: {iris.feature_names}")
```

**Dataset shape:** 150 samples and 4 features. Features are flowers characteristics

```python
#### Visualize original data (2D projection) ### 

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)

for i in range(3):
    plt.scatter(X[i,0], X[i,1], label=iris.target_names[i])  
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])
    plt.title("Original Data (First 2 Features)")
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```
