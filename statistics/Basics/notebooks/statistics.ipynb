### Bessel's correction

• Estimating the **population variance** from a sample and unknown population mean 

• Reduce the bias due to **finite sample size** 

• The residuals sum to 0, and we have $n-1$ independent residuals (d.o.f)

$f(n) = \frac{n}{n-1}$


```python
### Bessel's correction ### 

from matplotlib import pyplot as plt
import numpy as np

def f(n):
    """
    Bessel's correction function
    """
    b_factor = n / (n - 1) # add a warning to undefined input values
    return b_factor
```

```python
x = np.arange(2, 51) # generate integer values btw 2 and 50

print(f(300)) # 300 observations lead to 0.3% error
plt.plot(x, f(x))
plt.show()
```
